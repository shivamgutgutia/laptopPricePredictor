```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
%matplotlib inline
```

```python
df=pd.read_csv(r"C:\Users\shiva\Downloads\laptop_data.csv")
```

```python
df=df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
df["Ram"]=df["Ram"].str.replace("GB","")
df["Ram"]=df["Ram"].astype("int8")
df.head()
```

```python
df["Weight"]=df["Weight"].str.replace("kg","")
df["Weight"]=df["Weight"].astype("float16")
df.head()
```

```python
def convert(size):
    value,unit=size[:-2],size[-2:]
    if unit=="TB":
        return int(float(value))*1024
    else:
        return int(float(value))

def memorysplit(text):
    text=text.split()
    out=["0","0","0","0"]
    for i in range(len(text)):
        if text[i]=="HDD":
            out[0]=convert(text[i-1])
        elif text[i]=="SSD":
            out[1]=convert(text[i-1])
        elif text[i]=="Flash":
            out[2]=convert(text[i-1])
        elif text[i]=="Hybrid":
            out[3]=convert(text[i-1])
        else:
            pass
    return out

df[['HDD', 'SSD', 'Flash', 'Hybrid']] = pd.DataFrame(df['Memory'].apply(memorysplit).tolist(), index=df.index)

df[['HDD', 'SSD', 'Flash', 'Hybrid']].astype("int16")
            
```

```python
df.drop(columns=["Memory"],inplace=True)
df.head()
```

```python
def gpucompany(text):
    return text.split()[0]

df["gpuComp"]=df["Gpu"].apply(gpucompany)
df.head()
```

```python
df["Company"].value_counts()
```

```python
def company(text):
    if text in ["Dell","Lenovo","HP","ASUS","Acer","MSI","Toshiba"]:
        return text
    else:
        return "Others"

df["Company"]=df["Company"].apply(company)
        
```

```python
df.head()
```

```python
df["ScreenResolution"].value_counts()
```

```python
def screen(text):
    text=text.split()
    res=text.pop(-1).split("x")
    res=int(res[0])*int(res[1])

    touchscreen=1 if "Touchscreen" in text else 0

    ips=1 if "IPS" in text else 0

    if "HD" in text or "HD+" in text:
        type=""
        type = "Full" if "Full" in text else type
        type = "Quad" if "Quad" in text else type
        type  = "4K" if "4K" in text else type
    elif "Retina" in text:
        type="Retina"
    else:
        type="Unknown"

    return [res,touchscreen,ips,type]

df[['Resolution', 'TouchScreen', 'IPS', 'ScreenType']] = pd.DataFrame(df['ScreenResolution'].apply(screen).tolist(), index=df.index)
df.head()

```

```python
df.drop(columns=["ScreenResolution"],inplace=True)
```

```python
def cpucompany(text):
    text=text.split()
    company=text[0]

    ghz=float(text[-1].replace("GHz",""))

    return([company,ghz])

df[['cpuComp', 'ClockRate']] = pd.DataFrame(df['Cpu'].apply(cpucompany).tolist(), index=df.index)
df.head()
```
